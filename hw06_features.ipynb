### Вопрос №7

Обучите класс `CountVectorizer` из `scikit-learn`, используя колонку Description из датасета Two Sigma Connect: Rental Listing Inquires таким образом, чтобы в словаре было не более 1000 слов, не меняя прочие параметры по умолчанию. Какое слово соответствует индексу 165 в словаре обученного `CountVectorizer`?

- **building**
- apartment
- trust
- renthop

```python
import json
import pandas as pd 

with open('data/train.json', 'r') as raw_data:
    data =  json.load(raw_data)
    X_train = pd.DataFrame(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=1000)
vectorizer.fit_transform (X_train.description)
```

```python
for word, idx in vectorizer.vocabulary_.items():
    if idx == 165 :
        print ("Word with 165 index is : ", word )
```

```python
data = get_data()[0]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)

df = pd.DataFrame(df_scaled.apply(lambda x : np.std(x)) , columns = ['std'] )
low_std_features = df.index[df['std'] < 0.1].tolist()

data.drop(low_std_features, axis=1,  inplace=True)
print(len(data.columns))
```
